```python
import yfinance as yf
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as ssd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# =============================================================================
# 1. Data Download and Setup for Sector Data (Corrected)
# =============================================================================

# Define sector tickers
sector_tickers = [
    "^CNXIT", "^CNXAUTO", "^NSEBANK", "^CNXFMCG", "^CNXMEDIA",
    "^CNXMETAL", "^CNXPSE", "^CNXMNC", "^CNXINFRA",
    "^CNXREALTY", "^CNXENERGY", "^CNXPSUBANK", "^CNXSERVICE"
]

# Define the time period (last 5 years)
end_date = datetime.today()
start_date = end_date - timedelta(days=5 * 365)

# Download data for each sector and store Close prices (adjusted)
data_dict = {}
for ticker in sector_tickers:
    df = yf.download(
        ticker, 
        start=start_date, 
        end=end_date, 
        progress=False,
        auto_adjust=True  # Adjusted prices in 'Close' column
    )
    if not df.empty and 'Close' in df.columns:
        data_dict[ticker] = df['Close']

if not data_dict:
    raise ValueError("No sector data downloaded. Check ticker symbols or network connection.")

# Create DataFrame with proper alignment
prices = pd.concat(data_dict, axis=1).dropna()
returns = prices.pct_change().dropna()

# =============================================================================
# 2. HRP Weight Calculation Function (Numerically Stable Version)
# =============================================================================
def get_hrp_weights(returns_df):
    """
    Calculate HRP weights with numerical stability fixes.
    Steps:
      1. Compute the correlation matrix.
      2. Force symmetry and calculate the distance matrix.
      3. Convert to a condensed distance matrix and perform hierarchical clustering.
      4. Order tickers by cluster leaves and allocate weights recursively.
    """
    # Calculate correlation matrix
    corr = returns_df.corr()
    
    # Ensure symmetry: average with its transpose, and set diagonal to 1
    corr = 0.5 * (corr + corr.T)
    np.fill_diagonal(corr.values, 1.0)
    
    # Create distance matrix; use clip to avoid negative values
    dist = np.sqrt(np.clip(0.5 * (1 - corr), 0.0, 1.0))
    dist = 0.5 * (dist + dist.T)  # Force symmetry
    
    # Convert to condensed distance matrix
    condensed_dist = ssd.squareform(dist, checks=False)
    
    # Hierarchical clustering using Ward method
    link = sch.linkage(condensed_dist, method='ward')
    
    # Order tickers using dendrogram leaves
    sorted_indices = sch.leaves_list(link)
    ordered_tickers = returns_df.columns[sorted_indices]
    
    # Initialize equal weights and cluster sizes for each ticker
    weights = pd.Series(1, index=ordered_tickers)
    cluster_sizes = pd.Series(1, index=ordered_tickers)
    
    # Recursive allocation of weights
    for i in range(len(ordered_tickers) - 1):
        left, right = ordered_tickers[i], ordered_tickers[i+1]
        total_size = cluster_sizes[left] + cluster_sizes[right]
        
        weights[left] *= cluster_sizes[right] / total_size
        weights[right] *= cluster_sizes[left] / total_size
        
        cluster_sizes[left] = total_size
        cluster_sizes[right] = total_size
    
    return weights / weights.sum()

# =============================================================================
# 3. HRP Portfolio Rebalancing Simulation (Quarterly)
# =============================================================================

print("\n--- HRP-Weighted Quarterly Rebalancing with Top N Sector Selection ---\n")

# Simulate quarterly rebalancing
quarter_end_dates = pd.date_range(start=prices.index[0], end=prices.index[-1], freq='Q')

initial_capital = 100000
results_summary = []       # Summary simulation results for each n
all_quarterly_results = {} # Detailed quarterly results for each n

# Loop for different top-n sector selections (n = 3,4,...,10)
for n in range(3, 11):
    cumulative_value = initial_capital
    quarterly_returns = []  # To store each quarter's return
    quarterly_results = []  # Detailed results per quarter

    print(f"\n===== Simulating for Top {n} Sectors =====\n")
    
    for i, q_end in enumerate(quarter_end_dates):
        # Define rebalancing period
        q_start = prices.index[0] if i == 0 else quarter_end_dates[i-1] + pd.Timedelta(days=1)
        if q_start >= prices.index[-1]:
            break
        
        # Use data prior to q_start for training (~1 year)
        training_end = q_start - pd.Timedelta(days=1)
        training_prices = prices[prices.index <= training_end]
        if training_prices.empty:
            continue
        training_returns = training_prices.pct_change().dropna()
        if training_returns.empty:
            continue
        
        # Rank sectors by Sharpe ratio and select top n
        sharpe_scores = training_returns.mean() / training_returns.std()
        selected_sectors = sharpe_scores.nlargest(n).index.tolist()
        if len(selected_sectors) == 0:
            continue
            
        # Compute HRP weights using training data of selected sectors
        selected_training_returns = training_returns[selected_sectors]
        hrp_weights = get_hrp_weights(selected_training_returns)
        
        # Get prices for selected sectors in the current period
        period_prices = prices.loc[(prices.index >= q_start) & (prices.index <= q_end), selected_sectors]
        if period_prices.empty or len(period_prices) < 2:
            continue
        
        # Calculate period returns
        start_prices = period_prices.iloc[0]
        end_prices = period_prices.iloc[-1]
        sector_period_returns = (end_prices / start_prices) - 1
        
        # Calculate portfolio return and update cumulative value
        portfolio_return = np.dot(hrp_weights.loc[selected_sectors], sector_period_returns)
        cumulative_value *= (1 + portfolio_return)
        quarterly_returns.append(portfolio_return)
        
        # Store quarterly details
        quarterly_results.append({
            'quarter_start': q_start,
            'quarter_end': q_end,
            'selected_sectors': selected_sectors,
            'hrp_weights': hrp_weights.to_dict(),
            'portfolio_return': portfolio_return,
            'sector_returns': sector_period_returns.to_dict(),
            'cumulative_value': cumulative_value
        })
        
        # Print quarterly details
        print(f"Quarter {q_start.date()} to {q_end.date()}:")
        print(f"  Selected Sectors: {selected_sectors}")
        print("  HRP Weights:")
        for sector in selected_sectors:
            print(f"    {sector}: {hrp_weights[sector]:.2%}")
        print(f"  Portfolio Return: {portfolio_return:.2%}")
        print(f"  Cumulative Value: ₹{cumulative_value:,.2f}")
        print("-" * 80)
        
    annualized_vol = np.std(quarterly_returns) * np.sqrt(4) if len(quarterly_returns) > 1 else np.nan

    results_summary.append({
        'Number of Sectors': n,
        'HRP Portfolio Value': cumulative_value,
        'Annualized Volatility': annualized_vol
    })
    all_quarterly_results[n] = quarterly_results

hrp_results_df = pd.DataFrame(results_summary)
print("\n=== Simulation Summary for HRP Portfolios (Top N Sectors) ===")
print(hrp_results_df)

# =============================================================================
# 4. Fixed Benchmark Analysis with Scalar Values
# =============================================================================

benchmark_ticker = "^NSEI"
time_periods = {f"{i} Years": end_date - timedelta(days=365 * i) for i in range(1, 11)}
initial_capital_bench = 100000
benchmark_results = []

benchmark_data = yf.download(
    benchmark_ticker,
    start=min(time_periods.values()),
    end=end_date,
    progress=False,
    auto_adjust=True
)

if benchmark_data.empty:
    raise ValueError("No benchmark data downloaded. Check the ticker or network connection.")

benchmark_series = benchmark_data['Close'].squeeze()

for period, start_dt in time_periods.items():
    filtered_data = benchmark_series[benchmark_series.index >= pd.Timestamp(start_dt)]
    if len(filtered_data) < 2:
        continue
    
    start_price = float(filtered_data.iloc[0])
    end_price = float(filtered_data.iloc[-1])
    
    total_return = (end_price / start_price) - 1
    num_years = int(period.split()[0])
    cagr = ((1 + total_return) ** (1/num_years)) - 1  # Correct CAGR formula
    final_value = initial_capital_bench * (1 + total_return)
    
    benchmark_results.append({
        'Time Period': period,
        'Total Return (%)': round(total_return * 100, 2),
        'CAGR (%)': round(cagr * 100, 2),
        'Final Portfolio Value': round(final_value, 2)
    })

benchmark_results_df = pd.DataFrame(benchmark_results)
print("\n=== NIFTY 50 Benchmark Performance ===")
print(benchmark_results_df.to_string(index=False))

# =============================================================================
# 5. Final Comparison with Proper Data Types
# =============================================================================

bench_5yr = benchmark_results_df[benchmark_results_df["Time Period"] == "5 Years"]
if bench_5yr.empty:
    raise ValueError("5-Year benchmark data not available.")
nifty_5yr_final = float(bench_5yr.iloc[0]['Final Portfolio Value'])

comparison_df = hrp_results_df.copy()
comparison_df['NIFTY Value'] = nifty_5yr_final
comparison_df['Outperformance (%)'] = (
    (comparison_df['HRP Portfolio Value'] - nifty_5yr_final) 
    / nifty_5yr_final 
    * 100
).round(2)

print("\n=== Final Performance Comparison ===")
print(comparison_df[['Number of Sectors', 
                    'HRP Portfolio Value', 
                    'NIFTY Value', 
                    'Outperformance (%)']].to_string(index=False))

plt.figure(figsize=(12, 6))
bars = plt.bar(
    comparison_df['Number of Sectors'], 
    comparison_df['Outperformance (%)'],
    color=np.where(comparison_df['Outperformance (%)'] > 0, 'g', 'r')
)
plt.title("HRP Portfolio vs NIFTY 50 Performance (5-Year Period)")
plt.xlabel("Number of Sectors")
plt.ylabel("Performance Difference (%)")
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(axis='y')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom')
plt.show()

# =============================================================================
# 7. Investment Advice for Today (Using Last 1 Year Data)
# =============================================================================

advice_date = datetime.today()
advice_training_end = advice_date - timedelta(days=1)
advice_training_start = advice_training_end - timedelta(days=252)
if advice_training_start < prices.index[0]:
    advice_training_start = prices.index[0]

advice_training_prices = prices.loc[advice_training_start:advice_training_end].ffill()
advice_training_returns = advice_training_prices.pct_change().dropna()

print(f"\n--- Investment Advice as of {advice_date.date()} ---\n")
print(f"Using training data from {advice_training_start.date()} to {advice_training_end.date()}:\n")

for n in range(3, 11):
    sharpe_scores = advice_training_returns.mean() / advice_training_returns.std()
    top_n_sectors = sharpe_scores.nlargest(n).index.tolist()
    if not top_n_sectors:
        continue
    selected_training_returns = advice_training_returns[top_n_sectors]
    hrp_weights = get_hrp_weights(selected_training_returns)
    print(f"--- Top {n} Sectors Recommendation ---")
    print("Selected Sectors:", top_n_sectors)
    print("Recommended HRP Weights:")
    for sector in top_n_sectors:
        print(f"  {sector}: {hrp_weights[sector]:.2%}")
    print("-" * 60)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as ssd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# =============================================================================
# Get user input for simulation years
# =============================================================================
while True:
    try:
        years = int(input("Enter the number of years for the simulation (e.g., 7): "))
        if years <= 0:
            raise ValueError("Number of years must be positive.")
        break
    except ValueError as e:
        print("Invalid input. Please enter a positive integer for the number of years.")

# =============================================================================
# 1. Data Download and Setup for Sector Data (Corrected)
# =============================================================================

end_date = datetime.today()
start_date = end_date - timedelta(days=years * 365)

# Define sector tickers
sector_tickers = [
    "^CNXIT", "^CNXAUTO", "^NSEBANK", "^CNXFMCG", "^CNXMEDIA",
    "^CNXMETAL", "^CNXPSE", "^CNXMNC", "^CNXINFRA",
    "^CNXREALTY", "^CNXENERGY", "^CNXPSUBANK", "^CNXSERVICE"
]

# Download data for each sector and store Close prices (adjusted)
data_dict = {}
for ticker in sector_tickers:
    df = yf.download(
        ticker, 
        start=start_date, 
        end=end_date, 
        progress=False,
        auto_adjust=True  # Adjusted prices in 'Close' column
    )
    if not df.empty and 'Close' in df.columns:
        data_dict[ticker] = df['Close']

if not data_dict:
    raise ValueError("No sector data downloaded. Check ticker symbols or network connection.")

# Create DataFrame with proper alignment
prices = pd.concat(data_dict, axis=1).dropna()
returns = prices.pct_change().dropna()

# =============================================================================
# 2. HRP Weight Calculation Function (Numerically Stable Version)
# =============================================================================
def get_hrp_weights(returns_df):
    """
    Calculate HRP weights with numerical stability fixes.
    Steps:
      1. Compute the correlation matrix.
      2. Clean and force symmetry; calculate the distance matrix.
      3. Convert to a condensed distance matrix and perform hierarchical clustering.
      4. Order tickers by dendrogram leaves and allocate weights recursively.
    """
    # Calculate correlation matrix and clean it up
    corr = returns_df.corr()
    # Replace infinite values and NaNs with 0, then force diagonal to 1
    corr = corr.replace([np.inf, -np.inf], np.nan).fillna(0)
    np.fill_diagonal(corr.values, 1.0)
    
    # Force symmetry by averaging with its transpose
    corr = 0.5 * (corr + corr.T)
    
    # Create distance matrix; use clip to avoid negatives
    dist = np.sqrt(np.clip(0.5 * (1 - corr), 0.0, 1.0))
    dist = 0.5 * (dist + dist.T)  # Force symmetry
    
    # Convert to condensed distance matrix
    condensed_dist = ssd.squareform(dist, checks=False)
    
    # Hierarchical clustering using Ward method
    link = sch.linkage(condensed_dist, method='ward')
    
    # Order tickers using dendrogram leaves
    sorted_indices = sch.leaves_list(link)
    ordered_tickers = returns_df.columns[sorted_indices]
    
    # Initialize equal weights and cluster sizes for each ticker
    weights = pd.Series(1, index=ordered_tickers)
    cluster_sizes = pd.Series(1, index=ordered_tickers)
    
    # Recursive allocation of weights
    for i in range(len(ordered_tickers) - 1):
        left, right = ordered_tickers[i], ordered_tickers[i+1]
        total_size = cluster_sizes[left] + cluster_sizes[right]
        
        weights[left] *= cluster_sizes[right] / total_size
        weights[right] *= cluster_sizes[left] / total_size
        
        cluster_sizes[left] = total_size
        cluster_sizes[right] = total_size
    
    return weights / weights.sum()

# =============================================================================
# 3. HRP Portfolio Rebalancing Simulation (Quarterly)
# =============================================================================

print("\n--- HRP-Weighted Quarterly Rebalancing with Top N Sector Selection ---\n")

quarter_end_dates = pd.date_range(start=prices.index[0], end=prices.index[-1], freq='Q')

initial_capital = 100000
results_summary = []       # Summary simulation results for each n
all_quarterly_results = {} # Detailed quarterly results for each n

for n in range(3, 11):
    cumulative_value = initial_capital
    quarterly_returns = []  # To store each quarter's return
    quarterly_results = []  # Detailed results per quarter

    print(f"\n===== Simulating for Top {n} Sectors =====\n")
    
    for i, q_end in enumerate(quarter_end_dates):
        q_start = prices.index[0] if i == 0 else quarter_end_dates[i-1] + pd.Timedelta(days=1)
        if q_start >= prices.index[-1]:
            break
        
        training_end = q_start - pd.Timedelta(days=1)
        training_prices = prices[prices.index <= training_end]
        if training_prices.empty:
            continue
        training_returns = training_prices.pct_change().dropna()
        if training_returns.empty:
            continue
        
        sharpe_scores = training_returns.mean() / training_returns.std()
        selected_sectors = sharpe_scores.nlargest(n).index.tolist()
        # Fix: Unwrap tickers if they are tuples
        selected_sectors = [s[0] if isinstance(s, tuple) else s for s in selected_sectors]
        if len(selected_sectors) == 0:
            continue
            
        selected_training_returns = training_returns[selected_sectors]
        hrp_weights = get_hrp_weights(selected_training_returns)
        
        period_prices = prices.loc[(prices.index >= q_start) & (prices.index <= q_end), selected_sectors]
        if period_prices.empty or len(period_prices) < 2:
            continue
        
        start_prices = period_prices.iloc[0]
        end_prices = period_prices.iloc[-1]
        sector_period_returns = (end_prices / start_prices) - 1
        
        portfolio_return = np.dot(hrp_weights.loc[selected_sectors], sector_period_returns)
        cumulative_value *= (1 + portfolio_return)
        quarterly_returns.append(portfolio_return)
        
        quarterly_results.append({
            'quarter_start': q_start,
            'quarter_end': q_end,
            'selected_sectors': selected_sectors,
            'hrp_weights': hrp_weights.to_dict(),
            'portfolio_return': portfolio_return,
            'sector_returns': sector_period_returns.to_dict(),
            'cumulative_value': cumulative_value
        })
        
        print(f"Quarter {q_start.date()} to {q_end.date()}:")
        print(f"  Selected Sectors: {selected_sectors}")
        print("  HRP Weights:")
        for sector in selected_sectors:
            print(f"    {sector}: {float(hrp_weights[sector]):.2%}")
        print(f"  Portfolio Return: {portfolio_return:.2%}")
        print(f"  Cumulative Value: ₹{cumulative_value:,.2f}")
        print("-" * 80)
        
    annualized_vol = np.std(quarterly_returns) * np.sqrt(4) if len(quarterly_returns) > 1 else np.nan

    results_summary.append({
        'Number of Sectors': n,
        'HRP Portfolio Value': cumulative_value,
        'Annualized Volatility': annualized_vol
    })
    all_quarterly_results[n] = quarterly_results

hrp_results_df = pd.DataFrame(results_summary)
print("\n=== Simulation Summary for HRP Portfolios (Top N Sectors) ===")
print(hrp_results_df)

# =============================================================================
# 4. Fixed Benchmark Analysis with Scalar Values
# =============================================================================

benchmark_ticker = "^NSEI"
# Use the same number of years entered by the user for benchmark analysis
time_periods = {f"{years} Years": end_date - timedelta(days=365 * years)}
initial_capital_bench = 100000
benchmark_results = []

benchmark_data = yf.download(
    benchmark_ticker,
    start=min(time_periods.values()),
    end=end_date,
    progress=False,
    auto_adjust=True
)

if benchmark_data.empty:
    raise ValueError("No benchmark data downloaded. Check the ticker or network connection.")

benchmark_series = benchmark_data['Close'].squeeze()

for period, start_dt in time_periods.items():
    filtered_data = benchmark_series[benchmark_series.index >= pd.Timestamp(start_dt)]
    if len(filtered_data) < 2:
        continue
    
    start_price = float(filtered_data.iloc[0])
    end_price = float(filtered_data.iloc[-1])
    
    total_return = (end_price / start_price) - 1
    num_years = int(period.split()[0])
    cagr = ((1 + total_return) ** (1/num_years)) - 1  # Correct CAGR formula
    final_value = initial_capital_bench * (1 + total_return)
    
    benchmark_results.append({
        'Time Period': period,
        'Total Return (%)': round(total_return * 100, 2),
        'CAGR (%)': round(cagr * 100, 2),
        'Final Portfolio Value': round(final_value, 2)
    })

benchmark_results_df = pd.DataFrame(benchmark_results)
print("\n=== NIFTY 50 Benchmark Performance ===")
print(benchmark_results_df.to_string(index=False))

# =============================================================================
# 5. Final Comparison with Proper Data Types
# =============================================================================

bench_period = f"{years} Years"
bench_data = benchmark_results_df[benchmark_results_df["Time Period"] == bench_period]
if bench_data.empty:
    raise ValueError(f"{bench_period} benchmark data not available.")
nifty_final = float(bench_data.iloc[0]['Final Portfolio Value'])

comparison_df = hrp_results_df.copy()
comparison_df['NIFTY Value'] = nifty_final
comparison_df['Outperformance (%)'] = (
    (comparison_df['HRP Portfolio Value'] - nifty_final) 
    / nifty_final 
    * 100
).round(2)

print("\n=== Final Performance Comparison ===")
print(comparison_df[['Number of Sectors', 
                    'HRP Portfolio Value', 
                    'NIFTY Value', 
                    'Outperformance (%)']].to_string(index=False))

plt.figure(figsize=(12, 6))
bars = plt.bar(
    comparison_df['Number of Sectors'], 
    comparison_df['Outperformance (%)'],
    color=np.where(comparison_df['Outperformance (%)'] > 0, 'g', 'r')
)
plt.title(f"HRP Portfolio vs NIFTY 50 Performance ({years}-Year Period)")
plt.xlabel("Number of Sectors")
plt.ylabel("Performance Difference (%)")
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(axis='y')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom')
plt.show()

# =============================================================================
# 7. Investment Advice for Today (Using Last 1 Year Data)
# =============================================================================

advice_date = datetime.today()
advice_training_end = advice_date - timedelta(days=1)
advice_training_start = advice_training_end - timedelta(days=252)
if advice_training_start < prices.index[0]:
    advice_training_start = prices.index[0]

advice_training_prices = prices.loc[advice_training_start:advice_training_end].ffill()
advice_training_returns = advice_training_prices.pct_change().dropna()

print(f"\n--- Investment Advice as of {advice_date.date()} ---\n")
print(f"Using training data from {advice_training_start.date()} to {advice_training_end.date()}:\n")

for n in range(3, 11):
    sharpe_scores = advice_training_returns.mean() / advice_training_returns.std()
    top_n_sectors = sharpe_scores.nlargest(n).index.tolist()
    # Fix: Unwrap tickers if they are tuples
    top_n_sectors = [s[0] if isinstance(s, tuple) else s for s in top_n_sectors]
    if not top_n_sectors:
        continue
    selected_training_returns = advice_training_returns[top_n_sectors]
    hrp_weights = get_hrp_weights(selected_training_returns)
    print(f"--- Top {n} Sectors Recommendation ---")
    print("Selected Sectors:", top_n_sectors)
    print("Recommended HRP Weights:")
    for sector in top_n_sectors:
        print(f"  {sector}: {float(hrp_weights[sector]):.2%}")
    print("-" * 60)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as ssd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# =============================================================================
# Get user input for simulation years
# =============================================================================
while True:
    try:
        years = int(input("Enter the number of years for the simulation (e.g., 7): "))
        if years <= 0:
            raise ValueError("Number of years must be positive.")
        break
    except ValueError as e:
        print("Invalid input. Please enter a positive integer for the number of years.")

# =============================================================================
# 1. Data Download and Setup for Sector Data (Corrected)
# =============================================================================

end_date = datetime.today()
start_date = end_date - timedelta(days=years * 365)

# Define sector tickers
sector_tickers = [
    "^CNXIT", "^CNXAUTO", "^NSEBANK", "^CNXFMCG", "^CNXMEDIA",
    "^CNXMETAL", "^CNXPSE", "^CNXMNC", "^CNXINFRA",
    "^CNXREALTY", "^CNXENERGY", "^CNXPSUBANK", "^CNXSERVICE"
]

# Download data for each sector and store Close prices (adjusted)
data_dict = {}
for ticker in sector_tickers:
    df = yf.download(
        ticker, 
        start=start_date, 
        end=end_date, 
        progress=False,
        auto_adjust=True  # Adjusted prices in 'Close' column
    )
    if not df.empty and 'Close' in df.columns:
        data_dict[ticker] = df['Close']

if not data_dict:
    raise ValueError("No sector data downloaded. Check ticker symbols or network connection.")

# Create DataFrame with proper alignment
prices = pd.concat(data_dict, axis=1).dropna()
returns = prices.pct_change().dropna()

# =============================================================================
# 2. HRP Weight Calculation Function (Numerically Stable Version)
# =============================================================================
def get_hrp_weights(returns_df):
    """
    Calculate HRP weights with numerical stability fixes.
    Steps:
      1. Compute the correlation matrix.
      2. Clean and force symmetry; calculate the distance matrix.
      3. Convert to a condensed distance matrix and perform hierarchical clustering.
      4. Order tickers by dendrogram leaves and allocate weights recursively.
    """
    # Calculate correlation matrix and clean it up
    corr = returns_df.corr()
    # Replace infinite values and NaNs with 0, then force diagonal to 1
    corr = corr.replace([np.inf, -np.inf], np.nan).fillna(0)
    np.fill_diagonal(corr.values, 1.0)
    
    # Force symmetry by averaging with its transpose
    corr = 0.5 * (corr + corr.T)
    
    # Create distance matrix; use clip to avoid negatives
    dist = np.sqrt(np.clip(0.5 * (1 - corr), 0.0, 1.0))
    dist = 0.5 * (dist + dist.T)  # Force symmetry
    
    # Convert to condensed distance matrix
    condensed_dist = ssd.squareform(dist, checks=False)
    
    # Hierarchical clustering using Ward method
    link = sch.linkage(condensed_dist, method='ward')
    
    # Order tickers using dendrogram leaves
    sorted_indices = sch.leaves_list(link)
    ordered_tickers = returns_df.columns[sorted_indices]
    
    # Initialize equal weights and cluster sizes for each ticker
    weights = pd.Series(1, index=ordered_tickers)
    cluster_sizes = pd.Series(1, index=ordered_tickers)
    
    # Recursive allocation of weights
    for i in range(len(ordered_tickers) - 1):
        left, right = ordered_tickers[i], ordered_tickers[i+1]
        total_size = cluster_sizes[left] + cluster_sizes[right]
        
        weights[left] *= cluster_sizes[right] / total_size
        weights[right] *= cluster_sizes[left] / total_size
        
        cluster_sizes[left] = total_size
        cluster_sizes[right] = total_size
    
    return weights / weights.sum()

# =============================================================================
# 3. HRP Portfolio Rebalancing Simulation (Quarterly)
# =============================================================================

# We'll suppress intermediate logging and only store final summary metrics.
results_summary = []       # Final results for each N (number of sectors)

# For each top-N selection (from 3 to 10)
for n in range(3, 11):
    cumulative_value = 100000  # initial capital
    quarterly_returns = []
    
    quarter_end_dates = pd.date_range(start=prices.index[0], end=prices.index[-1], freq='Q')
    
    for i, q_end in enumerate(quarter_end_dates):
        q_start = prices.index[0] if i == 0 else quarter_end_dates[i-1] + pd.Timedelta(days=1)
        if q_start >= prices.index[-1]:
            break
        
        training_end = q_start - pd.Timedelta(days=1)
        training_prices = prices[prices.index <= training_end]
        if training_prices.empty:
            continue
        training_returns = training_prices.pct_change().dropna()
        if training_returns.empty:
            continue
        
        sharpe_scores = training_returns.mean() / training_returns.std()
        selected_sectors = sharpe_scores.nlargest(n).index.tolist()
        # Unwrap tickers if they are tuples
        selected_sectors = [s[0] if isinstance(s, tuple) else s for s in selected_sectors]
        if len(selected_sectors) == 0:
            continue
            
        selected_training_returns = training_returns[selected_sectors]
        hrp_weights = get_hrp_weights(selected_training_returns)
        
        period_prices = prices.loc[(prices.index >= q_start) & (prices.index <= q_end), selected_sectors]
        if period_prices.empty or len(period_prices) < 2:
            continue
        
        start_prices = period_prices.iloc[0]
        end_prices = period_prices.iloc[-1]
        sector_period_returns = (end_prices / start_prices) - 1
        
        portfolio_return = np.dot(hrp_weights.loc[selected_sectors], sector_period_returns)
        cumulative_value *= (1 + portfolio_return)
        quarterly_returns.append(portfolio_return)
    
    annualized_vol = np.std(quarterly_returns) * np.sqrt(4) if len(quarterly_returns) > 1 else np.nan
    
    results_summary.append({
        'Number of Sectors': n,
        'HRP Portfolio Value': cumulative_value,
        'Annualized Volatility': annualized_vol
    })

hrp_results_df = pd.DataFrame(results_summary)

# =============================================================================
# 4. Fixed Benchmark Analysis with Scalar Values
# =============================================================================

benchmark_ticker = "^NSEI"
time_periods = {f"{years} Years": end_date - timedelta(days=365 * years)}
initial_capital_bench = 100000
benchmark_results = []

benchmark_data = yf.download(
    benchmark_ticker,
    start=min(time_periods.values()),
    end=end_date,
    progress=False,
    auto_adjust=True
)

if benchmark_data.empty:
    raise ValueError("No benchmark data downloaded. Check the ticker or network connection.")

benchmark_series = benchmark_data['Close'].squeeze()

for period, start_dt in time_periods.items():
    filtered_data = benchmark_series[benchmark_series.index >= pd.Timestamp(start_dt)]
    if len(filtered_data) < 2:
        continue
    
    start_price = float(filtered_data.iloc[0])
    end_price = float(filtered_data.iloc[-1])
    
    total_return = (end_price / start_price) - 1
    num_years_calc = int(period.split()[0])
    cagr = ((1 + total_return) ** (1/num_years_calc)) - 1  # Correct CAGR formula
    final_value = initial_capital_bench * (1 + total_return)
    
    benchmark_results.append({
        'Time Period': period,
        'Total Return (%)': round(total_return * 100, 2),
        'CAGR (%)': round(cagr * 100, 2),
        'Final Portfolio Value': round(final_value, 2)
    })

benchmark_results_df = pd.DataFrame(benchmark_results)

# =============================================================================
# 5. Final Comparison with Proper Data Types
# =============================================================================

bench_period = f"{years} Years"
bench_data = benchmark_results_df[benchmark_results_df["Time Period"] == bench_period]
if bench_data.empty:
    raise ValueError(f"{bench_period} benchmark data not available.")
nifty_final = float(bench_data.iloc[0]['Final Portfolio Value'])

comparison_df = hrp_results_df.copy()
comparison_df['NIFTY Value'] = nifty_final
comparison_df['Outperformance (%)'] = (
    (comparison_df['HRP Portfolio Value'] - nifty_final) 
    / nifty_final * 100
).round(2)

# Calculate additional metrics: Annualized Return (%) and Sharpe Ratio
comparison_df['Annualized Return (%)'] = (
    (comparison_df['HRP Portfolio Value'] / 100000) ** (1/years) - 1
) * 100
comparison_df['Sharpe Ratio'] = comparison_df['Annualized Return (%)'] / comparison_df['Annualized Volatility']

print("\n=== Final Performance Comparison ===")
print(comparison_df[['Number of Sectors', 
                     'HRP Portfolio Value', 
                     'NIFTY Value', 
                     'Outperformance (%)',
                     'Annualized Return (%)',
                     'Annualized Volatility',
                     'Sharpe Ratio']].to_string(index=False))

# =============================================================================
# 6. Decide the Best N Automatically Based on Highest Sharpe Ratio
# =============================================================================
best_row = comparison_df.loc[comparison_df['Sharpe Ratio'].idxmax()]
print(f"\nThe best performance is achieved with {int(best_row['Number of Sectors'])} sectors, "
      f"yielding a Sharpe Ratio of {best_row['Sharpe Ratio']:.2f}.")

# =============================================================================
# (Optional) Plot the Outperformance for the simulation period
# =============================================================================
plt.figure(figsize=(12, 6))
bars = plt.bar(
    comparison_df['Number of Sectors'], 
    comparison_df['Outperformance (%)'],
    color=np.where(comparison_df['Outperformance (%)'] > 0, 'g', 'r')
)
plt.title(f"HRP Portfolio vs NIFTY 50 Performance ({years}-Year Period)")
plt.xlabel("Number of Sectors")
plt.ylabel("Performance Difference (%)")
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(axis='y')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as ssd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# =============================================================================
# 1. HRP Weight Calculation Function (Numerically Stable Version)
# =============================================================================
def get_hrp_weights(returns_df):
    """
    Calculate HRP weights with numerical stability fixes.
    Steps:
      1. Compute the correlation matrix.
      2. Clean and force symmetry; calculate the distance matrix.
      3. Convert to a condensed distance matrix and perform hierarchical clustering.
      4. Order tickers by dendrogram leaves and allocate weights recursively.
    """
    corr = returns_df.corr()
    # Replace infinite values and NaNs with 0, then force diagonal to 1
    corr = corr.replace([np.inf, -np.inf], np.nan).fillna(0)
    np.fill_diagonal(corr.values, 1.0)
    
    # Force symmetry by averaging with its transpose
    corr = 0.5 * (corr + corr.T)
    
    # Create distance matrix; use clip to avoid negatives
    dist = np.sqrt(np.clip(0.5 * (1 - corr), 0.0, 1.0))
    dist = 0.5 * (dist + dist.T)  # Force symmetry
    
    # Convert to condensed distance matrix
    condensed_dist = ssd.squareform(dist, checks=False)
    
    # Hierarchical clustering using Ward method
    link = sch.linkage(condensed_dist, method='ward')
    
    # Order tickers using dendrogram leaves
    sorted_indices = sch.leaves_list(link)
    ordered_tickers = returns_df.columns[sorted_indices]
    
    # Initialize equal weights and cluster sizes for each ticker
    weights = pd.Series(1, index=ordered_tickers)
    cluster_sizes = pd.Series(1, index=ordered_tickers)
    
    # Recursive allocation of weights
    for i in range(len(ordered_tickers) - 1):
        left, right = ordered_tickers[i], ordered_tickers[i+1]
        total_size = cluster_sizes[left] + cluster_sizes[right]
        
        weights[left] *= cluster_sizes[right] / total_size
        weights[right] *= cluster_sizes[left] / total_size
        
        cluster_sizes[left] = total_size
        cluster_sizes[right] = total_size
    
    return weights / weights.sum()

# =============================================================================
# 2. Simulation over Multiple Periods with Transaction Costs
# =============================================================================
final_comparison_results = {}

# Define sector tickers (remains constant)
sector_tickers = [
    "^CNXIT", "^CNXAUTO", "^NSEBANK", "^CNXFMCG", "^CNXMEDIA",
    "^CNXMETAL", "^CNXPSE", "^CNXMNC", "^CNXINFRA",
    "^CNXREALTY", "^CNXENERGY", "^CNXPSUBANK", "^CNXSERVICE"
]

# Loop over simulation periods (years 1 to 10)
for sim_years in range(1, 11):
    print(f"\n=== Running Simulation for {sim_years} Year(s) ===\n")
    
    end_date = datetime.today()
    start_date = end_date - timedelta(days=sim_years * 365)
    
    # Download data for each sector for the given period
    data_dict = {}
    for ticker in sector_tickers:
        df = yf.download(
            ticker,
            start=start_date,
            end=end_date,
            progress=False,
            auto_adjust=True
        )
        if not df.empty and 'Close' in df.columns:
            data_dict[ticker] = df['Close']
    
    if not data_dict:
        print("No sector data downloaded for this period. Skipping...")
        continue
    
    prices = pd.concat(data_dict, axis=1).dropna()
    returns = prices.pct_change().dropna()
    
    # =============================================================================
    # HRP Portfolio Rebalancing Simulation (Quarterly) with Transaction Costs
    # =============================================================================
    quarter_end_dates = pd.date_range(start=prices.index[0], end=prices.index[-1], freq='Q')
    initial_capital = 100000
    results_summary = []  # Final results for each N
    prev = None         # To store previous weights for turnover calculation
    
    # Loop for each top-N selection (from 3 to 10)
    for n in range(3, 11):
        cumulative_value = initial_capital
        quarterly_returns = []
        prev = None  # Reset previous weights for each simulation of top-N
        
        for i, q_end in enumerate(quarter_end_dates):
            q_start = prices.index[0] if i == 0 else quarter_end_dates[i-1] + pd.Timedelta(days=1)
            if q_start >= prices.index[-1]:
                break
            
            training_end = q_start - pd.Timedelta(days=1)
            training_prices = prices[prices.index <= training_end]
            if training_prices.empty:
                continue
            training_returns = training_prices.pct_change().dropna()
            if training_returns.empty:
                continue
            
            sharpe_scores = training_returns.mean() / training_returns.std()
            selected_sectors = sharpe_scores.nlargest(n).index.tolist()
            # Unwrap tickers if they are tuples
            selected_sectors = [s[0] if isinstance(s, tuple) else s for s in selected_sectors]
            if len(selected_sectors) == 0:
                continue
            
            selected_training_returns = training_returns[selected_sectors]
            hrp_weights = get_hrp_weights(selected_training_returns)
            
            # Apply transaction costs if this is not the first rebalancing period
            if prev is not None:
                # Reindex previous weights to match current weights index
                prev_aligned = prev.reindex(hrp_weights[selected_sectors].index, fill_value=0)
                turnover = np.abs(hrp_weights[selected_sectors].values - prev_aligned.values).sum()
                # Transaction cost is 0.1% per buy and 0.1% per sell (0.2% total) on the turnover amount
                tx_cost = cumulative_value * turnover * 0.002
                cumulative_value -= tx_cost
            # Update previous weights for next period
            prev = hrp_weights[selected_sectors]
            
            period_prices = prices.loc[(prices.index >= q_start) & (prices.index <= q_end), selected_sectors]
            if period_prices.empty or len(period_prices) < 2:
                continue
            
            start_prices = period_prices.iloc[0]
            end_prices = period_prices.iloc[-1]
            sector_period_returns = (end_prices / start_prices) - 1
            
            portfolio_return = np.dot(hrp_weights[selected_sectors], sector_period_returns)
            cumulative_value *= (1 + portfolio_return)
            quarterly_returns.append(portfolio_return)
        
        annualized_vol = np.std(quarterly_returns) * np.sqrt(4) if len(quarterly_returns) > 1 else np.nan
        
        results_summary.append({
            'Number of Sectors': n,
            'HRP Portfolio Value': cumulative_value,
            'Annualized Volatility': annualized_vol
        })
    
        # Reset previous weights after finishing this simulation for N
        prev = None
        
    hrp_results_df = pd.DataFrame(results_summary)
    
    # =============================================================================
    # Fixed Benchmark Analysis with Scalar Values
    # =============================================================================
    benchmark_ticker = "^NSEI"
    time_periods = {f"{sim_years} Years": end_date - timedelta(days=365 * sim_years)}
    initial_capital_bench = 100000
    benchmark_results = []
    
    benchmark_data = yf.download(
        benchmark_ticker,
        start=min(time_periods.values()),
        end=end_date,
        progress=False,
        auto_adjust=True
    )
    
    if benchmark_data.empty:
        print("No benchmark data downloaded. Skipping benchmark comparison...")
        continue
    
    benchmark_series = benchmark_data['Close'].squeeze()
    
    for period, start_dt in time_periods.items():
        filtered_data = benchmark_series[benchmark_series.index >= pd.Timestamp(start_dt)]
        if len(filtered_data) < 2:
            continue
        
        start_price = float(filtered_data.iloc[0])
        end_price = float(filtered_data.iloc[-1])
        total_return = (end_price / start_price) - 1
        num_years_calc = int(period.split()[0])
        cagr = ((1 + total_return) ** (1/num_years_calc)) - 1
        final_value = initial_capital_bench * (1 + total_return)
        
        benchmark_results.append({
            'Time Period': period,
            'Total Return (%)': round(total_return * 100, 2),
            'CAGR (%)': round(cagr * 100, 2),
            'Final Portfolio Value': round(final_value, 2)
        })
    
    benchmark_results_df = pd.DataFrame(benchmark_results)
    
    # =============================================================================
    # Final Comparison with Proper Data Types
    # =============================================================================
    bench_period = f"{sim_years} Years"
    bench_data = benchmark_results_df[benchmark_results_df["Time Period"] == bench_period]
    if bench_data.empty:
        print(f"{bench_period} benchmark data not available.")
        continue
    nifty_final = float(bench_data.iloc[0]['Final Portfolio Value'])
    
    comparison_df = hrp_results_df.copy()
    comparison_df['NIFTY Value'] = nifty_final
    comparison_df['Outperformance (%)'] = (
        (comparison_df['HRP Portfolio Value'] - nifty_final) 
        / nifty_final * 100
    ).round(2)
    
    comparison_df['Annualized Return (%)'] = (
        (comparison_df['HRP Portfolio Value'] / 100000) ** (1/sim_years) - 1
    ) * 100
    comparison_df['Sharpe Ratio'] = comparison_df['Annualized Return (%)'] / comparison_df['Annualized Volatility']
    
    final_comparison_results[sim_years] = comparison_df

# =============================================================================
# Print Final Comparison Tables for Each Simulation Period
# =============================================================================
for sim_years in sorted(final_comparison_results.keys()):
    print(f"\n=== Final Performance Comparison for {sim_years} Year(s) ===")
    print(final_comparison_results[sim_years][['Number of Sectors', 
                                               'HRP Portfolio Value', 
                                               'NIFTY Value', 
                                               'Outperformance (%)',
                                               'Annualized Return (%)',
                                               'Annualized Volatility',
                                               'Sharpe Ratio']].to_string(index=False))

# Optionally, plot the Outperformance for a specific simulation period
selected_year = 4  # change as desired
if selected_year in final_comparison_results:
    comp_df = final_comparison_results[selected_year]
    plt.figure(figsize=(12, 6))
    bars = plt.bar(comp_df['Number of Sectors'], comp_df['Outperformance (%)'],
                   color=np.where(comp_df['Outperformance (%)'] > 0, 'g', 'r'))
    plt.title(f"HRP Portfolio vs NIFTY 50 Performance ({selected_year}-Year Period)")
    plt.xlabel("Number of Sectors")
    plt.ylabel("Performance Difference (%)")
    plt.axhline(0, color='black', linewidth=0.8)
    plt.grid(axis='y')
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.1f}%', ha='center', va='bottom')
    plt.show()
```
